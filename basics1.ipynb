```python
from qiskit import QuantumCircuit
import qiskit.quantum_info as qi
from qiskit.visualization import array_to_latex

state = qi.Statevector([0,1,0,0])
state.draw('latex')
```

```python
qc = QuantumCircuit(2)
qc.h(1)
qc.cx(1,0)
qc.draw()

fs = qi.Statevector.from_instruction(qc)
fs.draw('latex')

# Do you recognize this state?
# Maximally entangled state of two qubits. Bell state.
```

```python
fs = state.evolve(qc)
fs.draw('latex')
```

```python
rho = qi.DensityMatrix(fs)
rho.draw('latex')
```

```python
rho.draw('city')
```

```python
rho_red = qi.partial_trace(rho,[0])
rho_red.draw('city')
```

```python
S = qi.Operator.from_label('S')
array_to_latex(S)
```

```python
rho2 = rho_red.evolve(S)
rho2.draw('latex')
```
